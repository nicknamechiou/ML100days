# 檢視 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 開始檢視不同欄位是否有異常值

### 檢查申請貸款時的年齡 - DAYS_BIRTH

```python
print(app_train['DAYS_BIRTH'])
print('===========================================')
app_train['DAYS_BIRTH']/-365
```

```python
# DAYS_BIRTH: 客戶申請貸款時的年齡
(app_train['DAYS_BIRTH'] / (-365)).describe()
```

### 檢查申請貸款前，申請人已在現職工作的時間 - DAYS_EMPLOYED

```python
print(app_train['DAYS_EMPLOYED'])
print('=======================================')
print(app_train['DAYS_EMPLOYED'] / -365)
print('=======================================')
(app_train['DAYS_EMPLOYED'] / -365).describe()
```

```python
# DAYS_EMPLOYED: 申請貸款前，申請人已在現職工作的時間
plt.hist(app_train['DAYS_EMPLOYED'])
plt.show()
```

```python
## 查看每個獨有值的數量
app_train['DAYS_EMPLOYED'].value_counts()
```

### 從上面的圖與數值可以看出, 365243 顯然是個奇怪的數值

```python
## 過濾出所有'DAYS_EMPLOYED'等於365243的數據
##  異常數據
anom = app_train[app_train['DAYS_EMPLOYED'] == 365243]
anom
```

```python
## 其餘的正常數據
non_anom = app_train[app_train['DAYS_EMPLOYED'] != 365243]
non_anom
```

## 檢視異常數據中為我們的貸款目標(TARGET)占了多少(比例)

```python
## 由於TARGET裡面為0, 1，所以使用平均數計算，就能得到在異常數據中為我們的目標(TARGET == 1)占了多少
print(anom['TARGET'].mean())
## 百分比
print('The anomalies default on %0.2f%% loans' % (anom['TARGET'].mean() * 100))
```

## 檢視正常數據中為我們的貸款目標(TARGET)占了多少(比例)

```python
## 由於TARGET裡面為0, 1，所以使用平均數計算，就能得到在正常數據中為我們的目標(TARGET == 1)占了多少
print(non_anom['TARGET'].mean())
## 百分比
print('The non-anomalias default on %0.4f%% of loans' % (non_anom['TARGET'].mean() * 100) )
```

## 有多少筆的異常就業天數數據和其佔比 - 'DAYS_EMPLOYED'

```python
print('There are %d anomalous days of employment' % len(anom))
## 異常就業天數的數據佔整個數據集的比例
sum(app_train['DAYS_EMPLOYED'] == 365243) / len(app_train)
```

## 創建一個特徵欄位 - DAYS_EMPLOYED_ANOM 來紀錄 DAYS_EMPLOYED 是否異常

```python
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train.loc[:, ['DAYS_EMPLOYED','DAYS_EMPLOYED_ANOM']]
```

```python
print(app_train['DAYS_EMPLOYED_ANOM'].value_counts())
```

## 將DAYS_EMPLOYED中的異常值以 NaN來取代

```python
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)
app_train['DAYS_EMPLOYED']
```

## 視覺化 - Histogram

```python
app_train['DAYS_EMPLOYED'].plot.hist(title = 'Days Employment Histogram')
plt.xlabel('DAYS Employment')
plt.ylabel('Counts');
```

## 檢視 OWN_CAR_AGE 貸款人的車齡

```python
## 過濾出 OWN_CAR_AGE中不為 NaN的數據
print(app_train[~app_train.OWN_CAR_AGE.isnull()]['OWN_CAR_AGE'])
## 視覺化
plt.hist(app_train[~app_train.OWN_CAR_AGE.isnull()]['OWN_CAR_AGE'])
plt.title('Own Car Age')
plt.xlabel('Age')
plt.ylabel('Counts');
```

```python
## 檢視各車齡的數量
app_train['OWN_CAR_AGE'].value_counts()
```

```python
## 檢視車齡大於50年的數據數量
app_train[app_train['OWN_CAR_AGE'] > 50]['OWN_CAR_AGE'].value_counts()
```
