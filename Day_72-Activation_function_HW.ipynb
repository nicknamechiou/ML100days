# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
# day 72 前述流程 / python程式 對照
def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return 1 * (x > 0)

x = plt.linspace(-5,5,100)
```

```python
print(ReLU(x))
plt.plot(x, ReLU(x), 'r')
plt.show()
```

```python
print(dReLU(x))
plt.plot(x, dReLU(x), 'r')
plt.show()
```
